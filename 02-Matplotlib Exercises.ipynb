** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure(dpi=100)
ax1 =  fig.add_axes([0,0,0.9,0.9])
ax2 = fig.add_axes([0.1,0.5,0.3,0.2])

#ax1
ax1.plot(x, y, 'blue')
ax1.set_xlabel('X')
ax1.set_ylabel('z')
axes1.set_title('Zoom Out')

#ax2
ax2.plot(x, y, 'blue')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax2.set_xlim(20,22.0)
ax2.set_ylim(30,50)
```
